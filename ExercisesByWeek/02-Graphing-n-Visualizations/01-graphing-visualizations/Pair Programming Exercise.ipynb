#### Start by importing the necessary libaries and reading the text file into memory. Note: You'll need to use a different separator. Search in google for the documentation on how to do so.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

href = "http://www.dougstats.com/16-17RD.txt"
df = pd.read_fwf(href)
df.head(3)
```

#### What position has the most points? Bonus: Is this statistical significant difference?

```python
mn = df.groupby('PS')['PTS'].mean()
st = df.groupby('PS')['PTS'].std()
for nm, elm in zip(mn, st):
    print(nm, elm)
    
    
```

#### What player has the most number of rebounds?

```python
df.sort_values('TR', ascending=False)[['Player', 'TR']]
```

#### What team has the best three point conversion rate? ( made / attempts)

```python
df['3E'] = df['3M'] / df['3A']
```

```python
three_point_efficiency = df.groupby('Team').mean()[['3M', '3A']]
three_point_efficiency['3E'] = three_point_efficiency['3M'] / three_point_efficiency['3A']
three_point_efficiency.sort_values('3E')
```

#### What team has the most number of points?

```python
print(df.groupby('Team').sum()[['PTS']].sort_values('PTS', ascending=False))
```
